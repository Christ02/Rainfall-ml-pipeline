## Data leakage considerations
In the project, you want a model that reflects reality—never “cheating” by peeking at data you won’t have at prediction time. Let’s make sure your features only come from the past.

## Which Features Leak “Tomorrow’s” Answers?
- **rain_today** or **rainfall_amount_today** (they summarize the entire day’s rain)  
- **max_temp_today**, **min_humidity_today**, **avg_wind_speed_today**, etc.  
- Any feature that aggregates data from 00:00 to 23:59 **of the target day**


## A Cleaner Setup: Predict “Today” from “Up to Yesterday”

1. **New Target**: Predict whether **it will rain today**.  
2. **Allowed Features**: All weather metrics **up to and including yesterday**.

This matches real-world decisions—like “Should I take my bike to work today?”—and guarantees your model only sees information you’d truly have.

### Renaming for Clarity

Be explicit in your DataFrame so you never mix up targets:


```python
df = df.rename(columns={'RainToday': 'RainYesterday',
                        'RainTomorrow': 'RainToday'
                        })
```

## Extracting Seasonality

Weather is inherently seasonal—summer, autumn, winter and spring all behave differently. Let’s turn your `Date` column into a new `Season` feature and then drop `Date` (since `Season` captures the signal you care about). Here’s a simple way to do it:


### Create a function to map dates to seasons


```python
def date_to_season(date):
    month = date.month
    if (month == 12) or (month == 1) or (month == 2):
        return 'Summer'
    elif (month == 3) or (month == 4) or (month == 5):
        return 'Autumn'
    elif (month == 6) or (month == 7) or (month == 8):
        return 'Winter'
    elif (month == 9) or (month == 10) or (month == 11):
        return 'Spring'
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Apply the function to the 'Date' column
df['Season'] = df['Date'].apply(date_to_season)

# Drop the original Date column
df = df.drop(columns=['Date'])

df
```

We’ve got a solid feature set ready—time to build our model. But first, let’s peek at our target and see if it’s balanced. That way, we’ll know whether we need to tweak for any class imbalances before we dive into training.

```python
X = df.drop(columns='RainToday', axis=1)
y = df['RainToday']
```

```python
# Check how balanced the classes are
y.value_counts()
```

## Conclusions

- Rain occurs ~23.7% of days (1 791/7 557) → about **87 rainy days per year**  
- Always predicting “No” yields ~76.3% accuracy (5 766/7 557)  
- Dataset is imbalanced (≈3.2 No : 1 Yes)  
- Next steps:
  - Use a **stratified** train/test split  
  - Handle class imbalance (e.g. `class_weight='balanced'`, SMOTE)  
  - Evaluate with **F1 score** or **ROC–AUC** instead of raw accuracy 


### Split data into training and test sets, ensuring target stratification


```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

### Define preprocessing transformers for numerical and categorical features

```python
# Automatically detect numeric and categorical columns
numeric_features = X_train.select_dtypes(include=['number']).columns.tolist()
categorical_features = X_train.select_dtypes(include=['object', 'category']).columns.tolist()
```

### Define separate transformers for both feature types and combine them into a single preprocessing transformer


```python
# Scale the numeric features
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])

# One-hot encode the categoricals 
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

### Combine the transformers into a single preprocessing column transformer

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

### Create a pipeline by combining the preprocessing with a Random Forest classifier

```python
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])
```

### Define a parameter grid to use in a cross validation grid search model optimizer


```python
param_grid = {
    'classifier__n_estimators': [50, 100],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5]
}
```

### Pipeline usage in crossvalidation
Recall that the pipeline is repeatedly used within the crossvalidation by fitting on each internal training fold and predicting on its corresponding validation fold


## Perform grid search cross-validation and fit the best model to the training data
### Select a cross-validation method, ensuring target stratification during validation


```python
cv = StratifiedKFold(n_splits=5, shuffle=True)
```

### Instantiate and fit GridSearchCV to the pipeline

```python
grid_search = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring='accuracy',
    verbose=2
)

# Fit the grid search to the training data
grid_search.fit(X_train, y_train)
```

### Print the best parameters and best crossvalidation score


```python
print("\nBest parameters found: ", grid_search.best_params_)
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))
```

### Display your model's estimated score


```python
# Evaluate the best model on the test set
test_score = grid_search.score(X_test, y_test)
print("Test set score: {:.2f}".format(test_score))
```

Now that we’re hovering around 84% accuracy, let’s pull back the curtain on our best model and see where it’s getting tripped up.

### Get the model predictions from the grid search estimator on the unseen data

```python
# Generate predictions on the test set using the best found model
y_pred = grid_search.predict(X_test)
```

### Print the classification report

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

### Plot the confusion matrix 

```python
# Compute the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Create and plot the display
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

### Extract the feature importances

```python
# Extract feature importances from the trained Random Forest
feature_importances = grid_search.best_estimator_['classifier'].feature_importances_
```

Now let's extract the feature importances and plot them as a bar graph.


```python
# Combine numeric and categorical feature names
feature_names = numeric_features + list(
    grid_search.best_estimator_['preprocessor']
               .named_transformers_['cat']
               .named_steps['onehot']
               .get_feature_names_out(categorical_features)
)

# Extract importances
feature_importances = grid_search.best_estimator_['classifier'].feature_importances_

# Build the DataFrame before slicing
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

# Select top N features
N = 20
top_features = importance_df.head(N)

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(top_features['Feature'], top_features['Importance'])
plt.gca().invert_yaxis()  # Show most important on top
plt.title(f'Top {N} Most Important Features for Rain Prediction')
plt.xlabel('Importance Score')
plt.show()
```

### Swap in Logistic Regression

Now I’ll replace the Random Forest step in **my** pipeline with `LogisticRegression`, apply the provided `param_grid_lr`, and run `GridSearchCV` again. Once it finishes, I’ll compare the cross-validated scores of both models to choose the best one for predicting Melbourne’s rainfall.

```python
# Replace RandomForestClassifier with LogisticRegression
pipeline.set_params(classifier=LogisticRegression(random_state=42))

# Update the GridSearchCV object to use the new pipeline
grid_search.estimator = pipeline

# Define a new grid with Logistic Regression parameters
param_grid = {
    'classifier__solver': ['liblinear'],
    'classifier__penalty': ['l1', 'l2'],
    'classifier__class_weight': [None, 'balanced']
}

grid_search.param_grid = param_grid

# Fit the updated pipeline with LogisticRegression
grid_search.fit(X_train, y_train)

# Make predictions on the test set
y_pred = grid_search.predict(X_test)
```

###  Compare the results to your previous model.
Display the clasification report and the confusion matrix for the new model and compare your results with the previous model.


```python
print(classification_report(y_test, y_pred))

# Generate the confusion matrix 
conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure()
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d')

# Set the title and labels
plt.title('Rainfall Classification Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

# Show the plot
plt.tight_layout()
plt.show()
```
